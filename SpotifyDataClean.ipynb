```python
from SpotifyAPICall import SpotifyPlaylistSongExtract
import pandas as pd
```

```python
Playlist1 = '4gBVIsAsisBIZU69NkhxHu?si=23469748a4404f9d'
Playlist2 = '74xvZytywpvCU1IgcFCDf0?si=69d1ca458cfe465b'
```

```python
df1 = SpotifyPlaylistSongExtract(Playlist1)
df2 = SpotifyPlaylistSongExtract(Playlist2)
```

```python
df3 = df1.append(df2)
```

```python
df3.head()
```

```python
df3['Artist'].unique()
```

```python
df3['Artist'][df3['Song'] == 'Get Down Saturday Night'] = 'Oliver Cheatham'
df3['Artist'][df3['Song'] == 'HONEY (MR)'] = 'J.Y. Park'
df3['Artist'][df3['Song'] == 'Super Star'] = 'Eddy Kim'
df3['Artist'][df3['Song'] == 'Bodyguard'] = 'Hui'
df3['Artist'][df3['Song'] == 'RESPECT'] = 'Loco'
df3['Artist'][df3['Song'] == 'Day Day'] = 'BewhY'
```

```python
Western_Artist_List = ['Bruno Mars', 'Michael Jackson', 'The Jacksons', 'Robin Thicke',
       'Earth, Wind & Fire', 'Kool & The Gang', 'McFadden & Whitehead',
       'Marvin Gaye', 'The Brothers Johnson', 'R. Kelly', 'Prince',
       'The Whispers', 'Cheryl Lynn', 'Patrice Rushen',
       'Chris Brown', 'Pitbull', "The O'Jays", 'James Brown',
       'Olly Murs', 'DNCE', 'The Isley Brothers', 'Nelly','Shakira']
```

```python
def CreateRegion(row):
    if row['Artist'] in Western_Artist_List:
        Region = 'Western'
    elif row['Song'] == 'Get Down Saturday Night':
        Region = 'Western'
    else:
        Region = 'Eastern'
    return Region

df3['Region'] = df3.apply(CreateRegion, axis=1)
```
